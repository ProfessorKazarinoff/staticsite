## Plotting trig functions using matplotlib and Python

Plotting is an essential skill for engineers. Plots can reveal trends in data and outliers. Plots are a way to visually communicate results with your engineering team, supervisors and customers.  In this post we are going to plot a couple of trig functions using Python and **matplotlib**. Matplotlib is a plotting library that can produce line plots, bar graphs, histograms and many other types of plots. Matplotlib is not included in the standard library. If you downloaded python from python.org, you will need to install **matplotlib** with **pip** on the command line.

 We are going to build a couple of plots which show the trig functions **sin** and **cos**. We'll start by importing matplotlib and numpy using the standard lines ```import matplotlib.pyplot as plt``` and ```import numpy as np```. This means we can use the short alias ```plt``` and ```np``` when we call these two libraries.  You could ```import numpy as wonderburger``` and use ```wonderburger.sin``` to call numpy functions, but this would look funny to other engineers.  The line ```import numpy as np``` has become a common convention and look familiar to other engineers using Python.  In case you are working in a Juypiter notebook, the ```%matplotlib inline``` command is also necessary to view the plots directly in the notebook.

````python
import matplotlib.pyplot as plt
import numpy as np
# if using a jupyter notebook
%matplotlib inline    
````

Next we will build a set of x values to use in our plot. We'll do this in a Python list.

````python
x = np.arange(0,4*np.pi,0.1)   # start,stop,step
y = np.sin(x)
````

````python
plt.plot(x,y)
plt.show()
````

````python
x = np.arange(0,4*np.pi,0.1)   # start,stop,step
y = np.sin(x)
z = np.cos(x)
````

````python
plt.plot(x,y,x,z)
plt.show()
````

````python
x = np.arange(0,4*np.pi-1,0.1)   # start,stop,step
y = np.sin(x)
z = np.cos(x)
````

````python
plt.plot(x,y,x,z)
plt.xlabel('x values from 0 to 4pi')   # string must be enclosed with quotes '  '
plt.ylabel('sin(x) and cos(x)')
plt.title('sin(x) and cos(x) from 0 to 4pi')
plt.legend(['sin(x)','cos(x)'])        # legend entries as seperate strings in a list
plt.show()
````
