```python
df = pd.read_csv('stress_strain_data.csv', sep=',', header=None, skiprows=0)
df.columns = ['strain', 'stress']
df.head()
```

```python
df.plot()
```

### Using matplotlib

```python
strain = df['strain']
stress = df['stress']
plt.plot(strain,stress)
plt.xlabel('strain')
plt.ylabel('stress')
plt.title('matplotlib')
plt.show()
```
