```python
df = pd.read_excel('T574.xls', usecols=[4,5,6])
df.head()
```

```python
A0 = 3.14*(0.506/2)**2
df['stress'] = (df['Force']/A0)*0.001
df['strain'] = df['lsr_deltaL']*0.01
df.head()
```

```python
df1 = df[['strain', 'stress']]
df1.head()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.plot(df1['strain'],df1['stress'])
```
